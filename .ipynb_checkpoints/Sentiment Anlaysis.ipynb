```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

plt.style.use('ggplot')

import nltk
```

```python
# Read in data
# df = pd.read_csv('../input/amazon-fine-food-reviews/Reviews.csv')

# Read in tsv
df = pd.read_csv('Smart_Watch_Reviews.csv',sep=',')

print(df.shape)
# insert ID column for later use

df.insert(0, 'id', range(1, 1 + len(df)))
```

```python
example = df['Review'][2]
print(example)
```

```python
# Tokenize the review
# NLP needs to tokenize text to parse text for analysis

tokens = nltk.word_tokenize(example)
tokens[:10]
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
# Part of Speech tagging

tagged = nltk.pos_tag(tokens)
tagged[:10]
```

```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# Put tagged parts of speech into ENTITIES
# NLTK Chunk  //  ne_chunk()

entities = nltk.chunk.ne_chunk(tagged)
entities.pprint()
```

VADER Sentiment Analysis

```python
# DOWNLOAD VADER
from nltk.sentiment import SentimentIntensityAnalyzer

# PROGRESS BAR TRACKER
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
# Example of Sentiment Score // Positive

# Compound Score = aggregation of neg, neutral, pos  //  Range [-1.0 : 1.0]

sia.polarity_scores('I am so happy!')
```

```python
# Example of Sentiment Score // Negative
sia.polarity_scores('This is the worst thing ever.')
```

```python
# Run SIA on example
print('example =',example,'\n')
sia.polarity_scores(example)
```

```python
# Run the polarity score on the entire dataset
# create Result dictionary res{}

res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Review']
    myid = row['id']
    res[myid] = sia.polarity_scores(text)
```

```python
# Convert dictionary to dataframe

# dataframe is oriented wrong way, run .T to pivot table
vaders = pd.DataFrame(res).T

# Merge onto original dataframe
vaders = vaders.reset_index().rename(columns={'index': 'id'})

# left merge
vaders = vaders.merge(df, how='left')
```

```python
# Now we have sentiment score and metadata
vaders.head()
```

```python
# seaborn barplot to view VADERS

ax = sns.barplot(data=vaders,x='Rating', y='compound')
ax.set_title('Compund Score by Amazon Star Review')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.show()
```

```python
# Visualize positive, neutral, and negative SIA scores respectively

# use matplot .subplots to make a 1 x 3 grid of distinct score type

# RESULT: positivity score increase with star, neutral is flat, negativity decreases with star

fig, axs = plt.subplots(1, 3, figsize=(12, 3))
sns.barplot(data=vaders, x='Rating', y='pos', ax=axs[0])
sns.barplot(data=vaders, x='Rating', y='neu', ax=axs[1])
sns.barplot(data=vaders, x='Rating', y='neg', ax=axs[2])


axs[0].set_title('Positive')
axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=90) 
axs[1].set_title('Neutral')
axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=90) 
axs[2].set_title('Negative')
axs[2].set_xticklabels(axs[2].get_xticklabels(), rotation=90) 

# RUN   plt.tight_layout()  to fix overlapping yaxis labels
plt.tight_layout()


plt.show()
```
